```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
df = sns.load_dataset('diamonds')
df
```

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

```python

sns.set({'figure.figsize':(13,13)})
palette = "ch:r=-.2,d=.3_r"

sns.scatterplot(
    data=df,
    x="carat", y="price",
    hue="clarity", size="depth", sizes = (1,10),palette = "ch:r=-.2,d=.3_r"
)

plt.show()



#code here
```

# Soal 2: Catplot

Buatlah sebuah Categorical plot dengan kind = 'boxplot', sehingga diperoleh hasil seperti expected output berikut:


```python
#code here


with sns.axes_style(style='ticks'):
    g = sns.catplot( data=df, x = "clarity", y =  "price", col = "cut", kind='box', showfliers=False)
    g.set_axis_labels("Clarity", "Price");
    

plt.show()
```

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price serta markers =  ['p', 'o', 's', 'D', 'X'] sehingga dihasilkan output sesuai expected.

```python
df = df.sample(n=300, random_state=123)
with sns.axes_style('darkgrid'):
      sns.pairplot(data = df, vars = ['carat', 'depth', 'table', 'price'], hue= 'cut', markers=["p", "o", "s", "D", "X"], corner=True)
#code here

plt.show()
```

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

```python
#code here
df = df.sample(n=300, random_state=123)
with sns.axes_style('white'):
    sns.jointplot(data=df, x = "carat", y = "price", kind='kde');
  
```

```python
df = df.sample(n=300, random_state=123)
with sns.axes_style('darkgrid'):

    sns.jointplot(data=df, x = "carat", y = "price", kind='reg') 
```
