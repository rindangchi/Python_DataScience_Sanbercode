<h2>Soal 2: Membuat Component Figure dan Axis</h2>

- Buatlah Figure dan Axis kosong tanpa Data
- Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit

```python
import numpy as np
import matplotlib.pyplot as plt


#Buatlah Figure dan Axis kosong tanpa Data
# membuat objek figure
fig = plt.figure()

# membuat objek axes
ax = fig.add_subplot()

# menampilkan objek visual
plt.show()


#Plot data di bawah ini dengan 2 cara, yaitu dengan membuat figure dan axis secara explicit dan Implicit
#cara explicit
x = np.linspace(0, 20, 100)
y = np.cos(x/2)
ax.plot(x,y)
plt.show()


#cara implisit
plt.plot(x, y)
plt.show()
```

```python
#dengan cara eksplisit
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 20, 100)
y = np.cos(x/2)

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x,y)
plt.show()


```

```python
#dengan cara implisit : tidak perlu mendefinisikan figure, figure sudah didefinikan oleh python 
plt.plot(x,y)
plt.show()
```
