```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, sharex=True, sharey=True, figsize= (18,8))

ax1.plot(x, tan, color = 'blue', label = 'tanX' )
ax2.plot(x, sin, color = 'red',  label = 'sinX')
ax3.plot(x, cos, color = 'green',  label = 'cosX')
fig.legend(loc = 'upper center')

ax1.set_title('Grafik Tanx')
ax2.set_title('Grafik Sinx')
ax3.set_title('Grafik Cosx')


fig.show()


```

```python
#session with mentor
fig, ax = plt.subplots(1, 3, figsize=(20,8), sharex=True, sharey=True)
ax[0].plot(x, tan, 'b', label = 'tan X')
ax[1].plot(x, sin, 'r', label = 'sin X')
ax[2].plot(x, cos, 'g', label = 'cos X')

ax[0].set_title('Grafik Tanx')
ax[1].set_title('Grafik Sinx')
ax[2].set_title('Grafik Cosx')

fig.legend(loc = 'upper center')
fig.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
fig, ax = plt.subplots(figsize=(12,8))




ax.plot(x, y)
ins1 = ax.inset_axes([-0.13, 0.9, 0.33, 0.3])
ins1.plot(x, y)
ins1.plot(x, y5, color = 'yellow')

ins2 = ax.inset_axes([0.87, 0.9, 0.33, 0.3])
ins2.plot(x, y)
ins2.plot(x, y4,  color = 'black')

ins3 = ax.inset_axes([0.87, -0.13, 0.33, 0.3])
ins3.plot(x, y)
ins3.plot(x, y3,  color = 'green')


ins4 = ax.inset_axes([-0.13, -0.13, 0.33, 0.3])
ins4.plot(x, y)
ins4.plot(x, y2,  color = 'red')

ax.set_title('Grafik Cosx')
ins1.set_title('Grafik Cosx^5')
ins2.set_title('Grafik Cosx^4')
ins3.set_title('Grafik Cosx^3')
ins4.set_title('Grafik Cosx^2')

plt.show()
```

```python
#session with mentor
fig, ax = plt.subplots(figsize = (14,8))

ax.plot(x, y)
ax.set_title('Grafik Cosx')

ins1 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins1.plot(x, y2, 'r', x, y)
ins1.set_title('Grafik cos^2x')

ins2 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins2.plot(x, y3, 'g', x, y)
ins2.set_title('Grafik cos^3x')

ins2 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins2.plot(x, y4, 'black', x, y)
ins2.set_title('Grafik cos^3x')

ins2 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins2.plot(x, y4, 'y', x, y)
ins2.set_title('Grafik cos^3x')

plt.show()
```
